# <리스트의 동적 배열>

리스트는 어떻게 배열과 다르게 크기의 구애를 받지 않을까?→  필요한 양보다 항상 넉넉한 크기의 메모리를 사용하기 때문

Ex. 크기가 3인 자료를 넣을거라 용량이 3인 리스트만 필요해도 용량이 10정도인 리스트를 할당하는 것

만약 자료를 점점 넣다가 리스트가 FULL상태가 되면 어떻게 할까?→ 파이썬은 동적 배열을 사용→ 추가적인 공간이 필요하면, 기존에 참조하던 메모리에서 더 큰 새로운 메모리를 할당해 참조→ 이후 이전 메모리에 저장된 모든 항목들을 복사해서 새로운 메모리로 옮기기

→ 이처럼 리스트는 크기의 구애를 받지 않지만 메모리의 낭비가 발생→ 항상 추가로 삽입될 항목에 대한 메모리를 남겨놔야하기 때문

리스트의 시간 복잡도

1. append()→ 대부분의 경우 1(그냥 뒤에 넣으면 되니깐)→ But 리스트가 FULL이라면, n개의 자료들을 새롭게 복사해서 다른 메모리에 옮겨야 하기 때문에 O(n)
2. insert()→ O(n)→ 삽입을 할때는 삽입할 위치 이후의 인덱스들은 모두 한 칸씩 뒤로, 삽입할 위치 앞의 인덱스들은 한 칸씩 앞으로 옮겨야 하기 때문에 O(n)

```python
#배열을 이용한 리스트
capacity=100
array=[None]*capacity
size=0

def isEmpty():
    if size==0:
        return True
    else:
        return False

def isFull():
    return size==capacity

def getEntry(pos):
    if 0<=pos<size:#pos<=size가 되면, 
        return array[pos]
    else:
        return None

def insert(pos,e ):
    global size  #size는 전역변수
    if not isFull() and 0<=pos<=size:
        for i in range(size,pos,-1):
            array[i]=array[i-1]
        array[pos]=e
        size +=1
    else:
        print("리스트 overflow 또는 유효하지 않는 삽입 위치")
        exit()

def delete(pos):
    global size
    if not isEmpty() and 0 <= pos < size:
        e=array[pos]
        for i in range(pos,size-1):
            array[i]=array[i+1]
        size -=1
        return e
    else:
        print("리스트 underflow 또는 유효하지 않는 삭제 위치")
        exit()
```

하지만 위의 코드에서 불편한 점이 있다→ 지금은 리스트 1개를 통해 객체를 찍어내는데 만약 리스트를 여러개 만들어서 찍어야 한다면?→ 리스트 여러개의 코드를 쭉 적어놓고 그 다음 객체로 작업을 하면 시간이 너무 많이 걸림→ 이를 방지하고자 클래스를 활용

1. 필요한 클래스를 선언(이때 클래스명의 첫글자는 무조건 대문자)
2. 전역변수로 선언된 데이터는 멤버 변수로 지정→ __init__(여기에)
3. 일반 함수로 구현된 애들은 모두 클래스의 멤버 함수(메소드)로 만듦
4. 클래스의 인스턴스 변수(**init** 함수 박스에 있는 애들)을 멤버함수에서 사용할 때는 앞에 self를 붙임(멤버 함수를 쓸 때도!)

```python
class ArrayList:
    def __init__(self,size, capacity=100):
        self.capacity=capacity
        self.array=[None]*capacity
        self.size=0


    def isEmpty(self):
        if size==0:
            return True
        else:
            return False

    def isFull(self):
        return self.size==self.capacity


    def getEntry(self,pos):
        if 0<=pos<self.size:
            return self.array[pos]
        else:
            return None

    def insert(self,pos,e):
        if not self.isFull() and 0<=pos<=self.size:
            for i in range(self.size,pos,-1):
                self.array[i]= self.array[i-1]
            self.array[pos]=e
            self.size +=1
        else:
            pass

    def delete(self,pos,value):
        if not self.isEmpty() and 0<=pos<self.size:
            value=self.array[pos]
            for i in range(pos,self.size-1):
                self.array[i]=self.array[i+1]
            self.size -=1
            return value
        else: 
            pass
```

집합은 리스트와 유사하지만, 2가지 차이가 존재

1. 원소의 중복을 허용X→ 위의 리스트와 다르게 중복되는게 있는지 찾는 함수가 필요
2. 원소들 사이에 순서X→ 따라서, 리스트와 다르게 인덱스가 존재 X

1. contains→ 리스트는 중복이 허용되는 반면, 집합은 중복 허용 X→ 따라서, 중복이 있는지를 체크하는 함수를 만들어야 함
2. insert→ 리스트에서 insert를 할려면 삽입할 값과 인덱스가 필요 & 삽입위치 다음 인덱스의 값들을 한칸 씩 뒤로 밀어야 했다→ 집합은 리스트와 다르게 원소간 순서가 없음→ 인덱스가 X→ 그냥 맨 뒤에 추가하면 됨(어차피 순서가 없어서 맨 뒤에 넣었다고 해도 맨 앞이나, 중간 등으로 자유롭게 움직일 수 있음)
3. delete→ 집합의 insert와 다르게 delete는 삭제시 그 다음 값들을 한칸씩 땡겨와야 함→ 맨 뒤의 값만 삭제하면 불필요한 연산을 줄일 수 있음→ 삭제할 값이랑 맨 뒤에 위치한 값을 바꿔치기(어차피 순서가 상관X)→ 따라서 리스트의 delete는 삭제할 인덱스 위치가 필요한 반면, 집합의 delete는 삭제할 값이 필요
