스택의 추상자료형

데이터: 후입선출의 접근 방법을 유지하는 항목들의 모음(자료형은 어떤 자료형이든 다 올 수 있음)

push(e) -> 요소 e를 스택의 맨 위에 추가-> 삽입과 같은 말
pop()->스택의 맨 위에 있는 요소를 꺼내 반환-> 비복원추출이기 때문에 값이 손실

peek()-> 스택의 맨 위에 있는 항목을 삭제하지 않고 반환-> 복원 추출
isEmpty()-> 스택이 비어있으면 True, 아니면 False를 반환
isFull()-> 스택이 가득 차 있으면 True, 아니면 False를 반환

스택에 연산을 적용하다보면 2개의 케이스와 마주하게 된다

1. 스택이 이미 FULL상태인데 거기에 push를 하려고 할때-> 당연히 입력이 불가능→called stack overflow
2. 스택이 이미 empty상태인데 거기에 pop()이나 peek()를 하려고 할때-> 당연히 추출이 불가능> called stack underflow

배열을 이용한 스택의 구조는 크게 2가지가 필요-> 요소를 저장할 배열 & 상단 요소의 위치(=인덱스)를 나타내는 변수

아래 코드에선 배열의 이름을 array, 상단 요소 위치를 나타내는 변수로 top, 배열의 용량은 capacity로 고정→
point is  top=capacity-1→ 만약 array에 3개의 요소가 있다면 이때의 top은 2(0,1,2-3개) -> stack=Empty-> -1(1개가 들어오면 0) stack=FULL-> capacity-1

```python
capacity=10 #capacity를 10으로 고정한다고 가정
array=[None]*capacity #array라는 배열을 capacity용량만큼 None으로 초기화
top=-1# None으로 초기화 했으니 당연히 top은 -1인 상태

#스택의 연산: 일반 함수
def isEmpty():
    
    if top==-1:
        return True
    else:
        return False
#비어있으면 당연히 상단이 -1이다        

def isFull():
    if top==capacity-1:
        return top==capacity-1
#꽉 차있으면 상단이 capacity-1이다(top==capacity X)-> top은 인덱스 값이고 인덱스는 0부터 시작하기 떄문

def push(e):
    global top#top은 전역변수
    if not isFull():
        top +=1
        array[top]=e
    else:
        print("stack overFlow")
        exit()

#먼저 FULL상태인지 체크해서 FULL상태이면 stack overflow를 출력하고 FULL상태가 아니면 상단을 1 올려주고 상단에 e를 추가한다
#한칸 올리고 값을 넣는다
def pop():
    global top
    if not isEmpty():
        top -=1
        return array[top+1]
    else:
        print("stack underflow")
        exit()
#EMPTY상태인지 체크해서 EMPTY상태이면 stack underflow를 출력하고, EMPTY상태가 아니면(값이 들어있으면) 상단을 1 줄여주고 상단 값을 반환
#이때 중요한 것은 array[top]이 아니라 array[top+1]임-> 왜냐면 return 하기 전에 top을 1 줄였기 때문에
#한칸 줄이고, 줄여진 값을 반환

def peek():
    if not isEmpty():
        return array[top]
    else:
        pass
```

From Function To Class!

```python
#ArrayStack.py

#배열로 구현된 스택을 함수가 아닌 클래스로 바꿔보자-> 직접 해보기
class Arraystack:

    def __init__(self,capacity):#생성자를 정의한다 이때 함수 부분의 전역 변수가 클래스의 멤버변수가 된다
        self.capacity=capacity
        self.Array=[None]*capacity#Array를 None으로 초기화 해준다
        self.top=-1#스택이 비어있을때를 상단은 -1이다-> 스택 역시 인덱스로 표시되므로 top=0이면 스택에 값이 1개가 저장되어 있다는 뜻으므로 -1로 설정해야 한다

    def isEmpty(self):
        if self.top==-1:#클래스의 멤버변수를 사용할 때는 변수앞에 self를 붙여준다
            return True#위에 서술했듯, 스택이 비어있으면 상단의 인덱스는 -1이다
        else:
            return False

    def isFull(self):
        if self.top==capacity-1:#반대로 스택이 가득차있으면 용량-1-> top==capacity이면 overflow가 발생
            return True
        else:
            return False


    def push(self,e):#push라는 함수의 지역변수 e는 클래스의 멤버변수가 아니기 때문에 앞에 self를 붙일 필요 X 
        if not isFull():
            self.top +=1#스택에 값을 삽입할때는, 먼저 상단의 인덱스를 한칸 올리고, 그 자리에 값을 삽입한다
            self.Array[top]=e
        else:
            print("Stack Overflow")#여기서 else인 경우는 stack이 가득찬 상태이므로 삽입불가
            exit()

    def pop(self): 
        if not isEmpty():#push와 반대로 pop은 최상단의 값을 추출하는 것
            self.top -=1#따라서 상단의 인덱스를 한칸 줄여주고 줄어든 인덱스에 위치한 값을 반환한다
            return self.Array[self.top+1]#이때 주의할 것은 push처럼 self.Array[top]으로 작성하면 안된다
#이미 앞에서 top을 한칸 줄여놨기 때문에 top에 +1을 해야 원래 값인 top의 인덱스가 되는 것-> 쉽게 생각하면 -1하고 +1해야 동일한 값이 되는 원리
        else:
            print("Stack Underflow")#여기서 else인 경우는 stack이 비어있는 상태이므로 추출 불가
            exit()

    def peek():#peek은 pop과 다르게 복원추출로 최상단의 값을 추출한다. 따라서, pop처럼 상단의 인덱스를 -1할 필요 X 그냥 꺼내면 됨
        if not isEmpty():
            return self.Array[self.top]
        else:
            print("Stack Underflow")
            exit()
```

→ 항상 클래스로 바꿀 때 주의할 점은 
1. 본래 함수의 전역번수는 클래스의 멤버변수로(**init**에 들어가는 애들), 함수 내에서 지역변수로 쓰이는 애들엔 self를 안붙이고, 멤버변수엔 self를 붙여야 한다

스택의 응용→ 1.  괄호 검사 알고리즘

괄호는 모든 언어에서 중요→ 괄호가 잘 사용되었는지 확인하는 알고리즘을 스택을 활용해 해결할 수 있음

정상적인 괄호의 조건 

1. 왼쪽 괄호의 개수=오른쪽 괄호의 개수
2. 왼쪽 괄호가 오른쪽 괄호보다 먼저 나와야 한다
3. 서로 다른 타입과 방향의 괄호가 서로 교차X→ 왼쪽 중괄호 왼쪽 소괄호가 나왔는데 갑자기 오른쪽 중괄호가 나오는…

→ 이를 스택으로 어떻게 해결할까?

1. 왼쪽 괄호가 나오면 해당 괄호를 스택에 집어넣는다
2. 오른쪽 괄호가 나오면 스택에서 pop연산으로 하나를 빼온다
3. 2번 과정에서 만약 빼올려고 했는데 스택이 비었거나, 다 빼왔는데 스택에 값이 남아있으면 Error

```python
#CheckBrackets.py

from Chapter4_소스코드.ArrayStack import ArrayStack

def CheckBrackets(statement):
    stack=ArrayStack(100)#CheckBrackets라는 함수를 만들고, stack이라는 변수에 용량이 100인 스택 객체 생성
    for ch in statement:
        if ch=='{' or ch=='(' or ch=='[':#ch라는 변수가 만약 [,(,{라는 왼쪽 괄호를 가지면
           stack.push(ch)#이를 stack이라는 객체에 넣는다

        elif ch=='}' or ch==')' or ch==']':#그렇지 않고 만약 ch가 ],),}라는 오른쪽 괄호를 가지면
            if stack.isEmpty():#우선 스택이 비었는지를 확인한다-> 스택이 비었는데 ch가 오른쪽 괄호를 가지면 왼쪽 괄호가 없어 짝이 안맞기 때문
                return False #스택이 빈 상태이면 괄호의 짝이 안맞기 때문에 False
            else:#스택이 빈 상태가 아니면=짝이 맞으면
                left=stack.pop()#스택에서 한개를 꺼내와 left에 저장
                if (ch =='}' and left != '{') or \
                (ch ==']' and left != '[') or \
                (ch ==')' and left != '('):#만약 꺼내온 괄호와 ch의 괄호가 다르면-> 괄호는 쌍을 이뤄야 하는데 이럴 경우 쌍을 이루지 못하기때문에 False
                    return False
    return stack.isEmpty()#괄호가 제 쌍을 다 찾아갔으면 스택이 공백
```

**2. 계산기 프로그램**

해당 알고리즘을 짜기 위해선 연산자 표기에 대한 인간과 컴퓨터의 표기방식의 차이를 알아야→ 인간은 중위 표기(피연산자1 연산자 피연산자2→ ex. 3+5=8, 4*2=8)를 선호→컴퓨터는 후위표기를 선호(피연산자1 피연산자2 연산자→ 3 5 +, 4 2 *)

컴퓨터는 왜 후위표기를 선호할까?

1. 괄호를 쓰지 않아도 계산 순서를 알 수 있음→  괄호를 고려하는건 컴퓨터 입장에선 계산비용이 발생하므로 손해

3. 수식을 읽으면서 동시에 연산 가능

하지만, 인간은 중위표기를 선호→  따라서 계산기 프로그램을 짜려면 일단 중위표기된 수식을 후위표기로 변환하는 과정부터 시작

3. 연산자가 나오면 스택에서 피연산자 2개를 꺼내 연산를 실행 & 실행결과를 다시 스택에 저장

그러면 중위표기된 수식을 어떻게 후위표기로 변환할까?

1. 연산자를 넣을 스택과 피연산자를 넣을 리스트를 만든다

Ex.

 A+B*C→ 이 경우 피연산자 A,B,C는 차례로 후위 표기식이라는 공백 리스트에 저장 & 연산자는 + * 순서로 스택에 삽입되는데, *이 +보다 연산순위가 높기 때문에 그대로 삽입

2. 괄호는 없지만 우선순위를 고려해야 하는 경우

A*B+C→ 이 경우를 1번 case처럼 그냥 넣을 경우 A*B+C가 아닌, A+B*C가 된다→ 따라서 이 경우 우선순위가 높은 연산자인 *은 모두 출력하고 그 뒤에 +연산자를 스택에 넣는다→ 

이를 일반화하면 1. 연산자를 스택에 넣을때 스택에 넣을 연산자보다 우선순위가 높거나 같은 연산자가 스택에 있다면 2.이를 모두 출력한 후 현재 연산자를 스택에 삽입

A*B+C→ A가 리스트에 저장→ *가 스택에 저장→ B가 공백 리스트에 저장→ +가 스택에 저장되어야 하는데 스택에 넣을 연산자인 +보다 이미 스택에 있는 *이 우선순위가 더 높음→ 따라서 *를 먼저 출력하고 +를 스택에 삽입→ 이후 C를 공백 리스트에 저장→ 그 뒤 스택에서 + 추출

3.  괄호가 있는 경우→ 괄호가 있는 수식은 반드시 우선 처리해야 하므로 이들을 위한 규칙이 추가

3-1. 왼쪽 괄호는 무조건 스택에 삽입하되, 우선순위가 가장 낮은 연산자로 취급

```python
#Infix2Postfix.py

from Chapter4_소스코드.ArrayStack import ArrayStack#ArrayStack 클래스 가져오기

def precedence(op):#연산자 간의 우선순위 결정
    if op == '(' or op == ')':
        return 0 #괄호는 우선순위를 제일 낮게 설정
    elif op == '+' or op == '-':
        return 1 #+와 -를 중간 우선순위로 설정
    elif op == '*' or op == '/':
        return 2 #*과 /를 최우선 우선순위로 설정
    else:
        return -1 # 연산자가 위의 것들에 해당하지 않으면 문제가 있는 것이므로 -1 
#이렇게 우선순위를 매기게 되면, 왼쪽 괄호가 스택에 들어가고 나서 오른쪽 괄호가 나오기 전까지 연산자들이 스택에 차례로 저장
#case2에서 볼 수 있듯 스택에 넣을 연산자보다 우선순위가 높은 애들이 스택에 있으면 이들을 모두 출력해야 하는데 괄호는 우선순위가 가장 낮기 때문에 오른쪽 괄호가 나오기 전까지 스택들이 차례로 저장될 수 있음

def Infix2Postfix(expr):
    s = ArrayStack(100)#s를 용량 100의 객체로 설정
    output = []# 중위 표기식을 후위 표기식으로 바꿀 때 이를 저장할 공백 리스트

    for term in expr:
        if term == '(':
            s.push('(')#중위식들을 term에 넣는데 term이 (이면 이를 스택에 저장

        elif term == ')':
            while not s.isEmpty():
                op = s.pop()
                if op == '(':
                    break
                else:
                    output.append(op)
'''
for term in expr:
        if term =='(':
            s.push('(')
            
        elif term ==')':
            if not s.isEmpty():
                X =s.pop()
                if X == '(':
                    break
                else:
                    result.append(X)

        elif term =='+' or term == '-' or term == '*' or term == '/':
            if s.isEmpty():
                s.push(term)
            else:
                X=s.pop()#있던 애
                if (precedence(X) > precedence(term)):#있던 애 > 넣을려는 애
                    s.push(term)
                else:
                    result.append(X)
                    s.push(term)
'''



#term에 )가 나오면 공백상태가 아닐때(underflow 방지) op라는 변수가 pop()된 스택의 연산자를 참조하도록 한 뒤
#후위 표기식에 이를 출력한다-> 해당 작업을 하다가 )가 나오면 이를 중지
#결과적으로 ( 연산자가 나오면 )가 나오기 전까지 스택에 있던 연산자들을 모두 출력하는 것
        elif term in '+-*/':
            while not s.isEmpty():
                op = s.peek()
                if precedence(term) <= precedence(op):
                    output.append(op)
                    s.pop()
                else:
                    break
            s.push(term)
#term에 +,-,*,/가 나오면, s가 공백상태가 아닐때 op변수가 peek()된 스택의 연산자를 참조하도록 한 뒤
#우선순위를 따져 op가 term보다 크면(=우선순위가 높으면) 우선순위가 높은 애를 후위표기식에 출력-> 그 뒤 s 객체에서 pop()작업
#우선순위가 높지 않을 때까지 해당 작업을 반복
#결과적으로 term이 연산자일때, term보다 우선순위가 높은 애들을 출력해서 후위 표기식에 저장하는 것
        else:
            output.append(term)#term이 (도 )도 +-*/도 아니면(=피연산자) 바로 후위표기식에 저장

    while not s.isEmpty():
        output.append(s.pop())#위의 작업들을 해도 객체가 비어있지 않으면, pop()해서 이를 후위 표기식에 저장

    return output
```

```python
#후위표기된 수식 연산
#EvalPostfix.py

from Chapter4_소스코드.ArrayStack import ArrayStack#객체를 찍을 클래스인 ArrayStack을 가져온다

def evalPostfix(expr):
    s = ArrayStack(100)#용량이 100인 s라는 객체를 찍는다
    for token in expr:
        if token in '+-*/':#만약 token이 +,-,*,/라면-> 즉, token이 연산자라면
            val2 = s.pop()#피연산자 2를 s라는 객체(스택)에서 하나 추출
            val1 = s.pop()#피연산자 1을 s라는 객체(스택)에서 하나 추출
            if (token == '+'):#이제는 각 연산자에 맞게 피연산자 1, 2를 연산해주어야 한다
                s.push(val1 + val2)#연산이 덧셈이면 피연산자 1,2를 더해주고 이를 삽입
            elif (token == '-'):
                s.push(val1 - val2)#연산이 뺄셈이면 피연산자 1,2를 빼고 이를 삽입
            elif (token == '*'):
                s.push(val1 * val2)#연산이 곱셈이면 피연산자 1,2를 곱한 결과를 삽입
            elif (token == '/'):
                s.push(val1 / val2)#연산이 나눗셈이면, 피연산자 1,2를 나눈 몫의 결과를 삽입
        else:#만약 token이 연산자가 아닌 피연산자라면
            s.push(float(token))#피연산자를 실수(float)로 변환해서 s라는 스택(객체)에 삽입
    
    return s.pop()
```

```python

from Chapter4_소스코드.ArrayStack import ArrayStack

map =[
		[ '1', '1', '1', '1', '1', '1' ],
	  [ 'e', '0', '0', '0', '0', '1' ],
	  [ '1', '0', '1', '0', '1', '1' ],
	  [ '1', '1', '1', '0', '0', 'x' ],
	  [ '1', '1', '1', '0', '1', '1' ],
	  [ '1', '1', '1', '1', '1', '1' ]]
MAZE_SIZE = 6

def isValidPos(x,y):#갈 수 있는 위치인지 파악하는 함수
    if 0<=x<MAZE_SIZE and 0<=y<MAZE_SIZE:#x와 y좌표가 map안에 있는 상태에서
        if map[y][x] == '0' or map[y][x]=='x': #-> 해당 위치가 0(갈 수 있는 칸)이거나 x(출구)라면
            return True #True를 출력
        return False
            
def DFS():#미로 출구를 찾는 함수
    print('DFS: ')
    stack=ArrayStack(100)
    stack.push((0,1))#행렬로 표현하면 map[1][0]. 좌표는 이를 뒤집은 (0,1)

    while not stack.isEmpty():
        here=stack.pop()
        print(here,end='->')
        (x,y)=here

        if(map[y][x] == 'x'):#x는 출구
            return True
        else:
            map[y][x]='.'
            if isValidPos(x,y-1):
                stack.push((x,y-1))#상-> 위로 한칸 갈려면 행을 하나 줄여야 함
            if isValidPos(x,y+1):
                stack.push((x,y+1))#하-> 아래로 한칸 갈려면 행을 하나 늘리면 됨
            if isValidPos(x-1,y):
                stack.push((x-1,y))#좌-> 왼쪽으로 한칸 갈려면 열을 하나 줄여야 함
            if isValidPos(x+1,y):
                stack.push((x+1,y))#우-> 오른쪽으로 한칸 갈려면 열을 하나 늘려야 함
        print('현재 스택',stack)
    return False
```
