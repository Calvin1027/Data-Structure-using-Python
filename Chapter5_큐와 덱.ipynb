원형 큐→ 배열을 원형으로 생각해 선형 큐의 단점을 보완→ 이때 주의할 것은 실제로 배열이 원형이 되는게 아니라, front와 rear의 인덱스를 시계방향(원형)으로 회전시킨다고 생각만 하는 것

시계방향은 어떻게 구현?→ front, rear를 증가시키다가 capacity와 같아지면, 다시 0으로 초기화→ if문보다는 %capacity로 한다
→ 어차피 front랑 rear의 인덱스는 capacity-1이므로 %capacity로 하면 절대 이들의 인덱스=capacity일수가 없다

1. 삽입→ rear의 인덱스를 시계방향으로 1만큼 회전→  rear=(rear+1)% capacity
2. 삭제→ front의 인덱스를 시계방향으로 1만큼 회전→ front=(front+1)% capacity

```python
class CircularQuere:
    def __init__(self,capacity=8):
        self.capacity=capacity
        self.array=[None]*capacity
        self.front=0
        self.rear=0


    def isEmpty(self):
        if self.front==self.rear:
            return True
        else:
            return False


    def isFull(self):
        if self.front==(self.rear+1)%self.capacity:#rear가 front바로 앞에 있으면 rear+1=front가 되고 front를 capacity로 나누면 front가 된다
            return True
        else:
            return False


    def enquence(self, value):
        if not isFull():
            self.rear=(self.rear+1)%self.capacity
            self.array[self.rear]=value
        else:
            print('삽입 불가-> overflow 의심')


    def dequeue(self):
        if not isEmpty():
            self.front=(self.front+1)%self.capacity
            return self.array[self.front]
        else:
            print('삭제 불가-> underflow 의심')


    def peek(self): #스택에선 맨 마지막, 큐에선 맨 앞 요소 복원추출
        if not isEmpty():
            return self.array[(self.front+1)%self.capacity]
        else:
            print('조회불가-> underflow 의심')


		def __str__(self): #이거 시험에 나올듯
				if self.front < self.rear:
						return str(self.array[self.front+1:self.rear+1])
				else:
						return str(self.array[self.front+1:self.capacity]+ self.array[0:self.rear+1])
```

1. front<rear→ front보다 rear가 오른쪽에 있는 경우→ 이때는 front+1부터 rear까지 슬라이싱을 이용해 출력
2. front>rear→ front가 rear보다 왼쪽에 있는 경우(rear가 기준점을 지나쳐 있는 경우)→ front+1부터 capacity-1까지, 0부터 rear까지를 출력

이때 주의할 것→ 

 만약 [front+1:rear]이렇게 슬라이싱을 하면 front+1부터 rear까지가 아니라 rear-1까지임!→ 슬라이싱의 마지막 부분 -1 한 곳 까지라는 것 기억하기!!
