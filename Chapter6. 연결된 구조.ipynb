<연결된 구조란?>

우리가 지금까지 구현한 자료구조(스택, 큐, 덱)는 모두 배열을 이용한 자료구조

배열의 가장 큰 특징은 모든 데이터를 연속된 메모리 공간에 저장하는 것→ 항상 용량을 넉넉하게 준비하기 때문에 낭비가 발생→ 이러한 낭비를 줄이기 위해 만들어진 것이 연결된 구조

→ 배열에 비해 데이터 탐색은 오래 걸리지만(처음부터 원하는 데이터까지 타고 가야하기 때문에) 삽입, 삭제가 자유롭고(링크만 바꾸면 되니까) 메모리의 낭비 X→ 결국 선 연결하는 시스템인데, 선을 연결하는 순서(삽입, 삭제)를 내가 알아서 정할 수 있다

연결된구조는 배열처럼 데이터를 한군데 모아두지 않고 데이터가 메모리의 여기저기에 다 흩어져 있다→ 
그럼 어떻게 흩어진 데이터를 모아서 관리할까?→ 데이터를 링크로 연결하여 표현!

<연결된 구조의 구성>

연결된 구조는 노드가 존재→ 노드는 데이터필드와 다음 노드로 갈 수 있는 링크필드를 가짐

데이터필드→ 우리가 저장하고 싶은 자료가 들어감→ 클래스의 객체, 다른 리스트…

링크필드→ 다른 노드를 가르키는(= 다른 노드의 주소를 저장하는 변수)

연결된 구조에서는 첫번째 노드만 알면 첫번째 노드의 링크를 타고 2번째 노드로, 2번째 노드를 타고 3번째 노드로… → 모든 노드에 접근할 수 있음

그런데 첫번째 노드는 어디를 타고가야할까?→ 당연히 타고갈 노드가 없다  따라서 첫번째 노드의 주소는 따로  저장해야 한다→ 연결 리스트에서 첫번째 노드의 주소를 저장하는 변수=헤드포인터

마지막 노드는 어떻게 끝내야 할까?→ 마지막 노드의 링크필드에 None을 저장하면 된다.

<단순 연결 구조의 응용- 연결된 스택>

스택을 배열로 표현하면 항상 등장했던게 index→ 연결구조에선 배열처럼 서로 딱 붙어 있지 않기 때문에 index는 필요없음→ 대신 다음 데이터가 있는 주소를 가져야 함

따라서 이 코드에서 top은 Index가 아니라, 첫번째 노드의 주소(링크)

````python
#노드 클래스 정의
class Node:
    def __init__(self,elem,link=None):
        self.data=elem #노드의 데이터
        self.link=link #다음 노드를 가르키는 링크

#연결된 스택
class LinkedStack:
    def __init__(self):
        self.top=None #초기 노드는 아무것도 없기에 다음 노드 역시 존재 X 따라서 헤드포인터인 top은 당연히 None를 가르켜야 한다

    def isEmpty(self):
        return self.top==None #위의 논리와 유사하게, 스택이 비어있으면 헤드포인터인 top이 None을 가르킨다

    def isFull(self):
        return False #연결된 구조는 용량제한이 없기 때문에 FUll상태가 존재하지 않는다 

		def push(self,e):
				self.top=Node(e,self.top)

    def pop(self):
        if not self.isEmpty():
            n=self.top 
            self.top=n.link 
            return n.data

    def peek(self):
        if not self.isEmpty():
            return self.top.data

		def size(self):
        node=self.top #node는 각 노드의 링크를 가르키는 변수
        count=0
        while not node==None:
            node=node.link #node가 현재 노드의 링크(다음 노드의 주소)를 가르키게 한다
            count +=1
        return count

    def __str__(self):
        arr=[]
        node=self.top
        while not node==None:
            arr.append(node.data)
            node=node.link
        return str(arr)
````

<단순 연결 구조의 응용- 연결된 리스트>

연결구조에서 가장 중요한 것은 첫번째 노드의 주소!

head=시작 노드를 가르키는 포인터(첫번째 노드의 주소를 가지고 있는 애)

````python
class LinkedList:
    def __init__(self):
        self.head=None

    def isEmpty(self):
        return self.head==None

    def isFull(self):
        return False

    def getNode(self,pos):
        if pos <0:
            return None #pos가 음수면 None

        node=self.head #node가 head를 가르키게 한다
        while pos >0 and node != None:
            node=node.link #node가 node의 링크(= 다음 노드가 있는 주소)를 가르킨다
            pos -= 1 #이렇게 계속 돌다가 pos가 0이 되면 끝
        return node


    def getEntry(self,pos):
        node=self.getNode(pos)
        if node==None:
            return None
        else:
            return node.data

    def insert(self,pos,elem):
        before=self.getNode(pos-1)
        if before ==None:
            self.head=Node(elem,self.head)
        else:
            node=Node(elem,before.link)
            before.link=node

    def delete(self,pos):
        before=self.getNode(pos-1)
        if before ==None:
            if self.head is not None:
                self.head=self.head.link
        elif before.link != None:
            before.link=before.link.link
````

def getNode(self,pos):
        if pos <0:
            return None #노드의 위치인 pos는 음수일 수 없기에 음수가 나오면 None으로 처리한다

<단순 연결 구조의 응용- 연결된 큐>

큐 역시 연결구조로 나타낼 수 있다

단순 선형 큐에선 front가 맨 앞의 노드를 rear가 맨 뒤에 노드를 가르킨다

하지만 우리가 자주 쓰는 선형 큐에선 마지막 노드(tail or rear)의 링크가 첫번째 노드를 가르키게 된다

따라서 이때는 front를 저장할 필요가 없다(tail.link=front이기 때문에)

````python
class Node:
    def __init__(self,elem,link=None):
        self.data=elem #노드의 데이터
        self.link=link #다음 노드를 가르키는 링크

class LinkedQueue:
    def __init__(self):
        self.tail=None #tail.link=front

    def isEmpty(self):
        if self.tail==None:
            return True

    def isFull(self):
        return False

    def enqueue(self,item):
        node=Node(item,None)
        if self.isEmpty():
            node.link=node
            self.tail=node
        else:
            node.link=self.tail.link #새로만든 node의 링크가 front를 가르키도록
            self.tail.link=node #원래 tail이었던 노드의 링크가 새로 만든 node를 가르키도록 한다
            self.tail=node #tail이 node를 가르키도록 한다

    def dequeue(self):
        if not self.isEmpty():
            data=self.tail.link.data
            if self.tail.link==self.tail:#큐에 1개가 있을 때
                self.tail=None
            else:
                self.tail.link=self.tail.link.link
            return data

    def size(self):
        if self.isEmpty():
            return 0
        else:
            count=1
            node=self.tail.link #front부터 시작
            while not node==self.tail:
                node=node.link
                count +=1
            return count

    def __str__(self):
        arr=[]
        if not isEmpty():
            node=self.tail.link
            while not node==self.tail:
                arr.append(node.data)
                node=node.link
            arr.append(node.data)#node가 계속 타고 가다가 self.tail이 되면 while문이 종료가 됨-> 이렇게 되면, 마지막 데이터는 arr에 담기지 않기 때문에 마지막 데이터까지 담아줘야 함
        return str(arr)
````

<이중 연결 구조의 응용-  연결된 덱>

왜 덱은 단순연결구조가 아닌 이중 연결구조를 사용할까?→ 이는 덱의 특성과 관련이 있다

전단에서는 삭제만, 후단에서는 삽입만 가능했던 큐와 다르게, 덱은 전단 후단 모두 삽입, 삭제가 가능하다

하지만, 연결된 덱에서 후단 삭제 연산을 진행할 때 문제가 발생한다.

배열에선 후단 삭제 연산을 할려면 현재 rear위치의 노드를 삭제하고 rear를 Rear-1을 해주었는데, 연결된 구조에선 선행노드를 알 수 없기 때문에 rear 전 노드의 위치를 모른다

따라서 이러한 문제를 해결하고자 선행노드의 정보를 포함하는 이중연결구조를 사용한다

이중연결구조는 단순연결구조와 다르게 링크가 1개가 아니라 2개가 된다(이전 노드의 링크, 이후 노드의 링크)

````python
class Dnode:
	def __init__(self,elem,prev=None,next=None):
		self.data=elem
		self.prev=prev
		self.next=next

class DoublyLinkedDeque:
    def __init__(self):
        self.front=None
        self.rear=None

    def isEmpty(self):
        if self.front==None:
            return True

    def isFull(self):
        return False

    def getfront(self):
        return self.front.data

    def getrear(self):
        return self.rear.data

    def addFront(self,item):
        node=Dnode(self,None,self.front) #전단삽입이니까 이전 노드=None
        if self.isEmpty():
            self.front=self.rear=node
        else:
            self.front.prev=node#front의 선행노드가 새로만든 node를 가르키게 한다
            self.front=node

    def addRear(self,item):
        node=Dnoe(self,self.rear,None)
        if self.isEmpty():
            self.rear=self.front=node
        else:
            self.rear.next=node
            self.rear=node

    def deletefront(self):
        if not isEmpty():
            data=self.front.data #전단에 삭제할 데이터 저장
            
            self.front=self.front.next
            if self.front==None: #이는 이 덱의 요소가 1개라는 뜻이고, 이 1개를 삭제하게 되니 덱은 공백상태가 되고 당연히 front랑 rear는 None을 가르키게 된다
                self.rear=None
            else: #요소를 지웠는데 덱이 공백상태가 아니라면
                self.front.prev=None #전단삭제기 때문에 맨 첫번째 노드의 prev는 None!
            return data

    def deleterear(self):
        if not isEmpty():
            data=self.rear.data #후단에 삭제할 데이터 저장
            self.rear=self.rear.prev#self.rear가 후단노드의 바로 직전 노드를 가르키게 한다
            if self.rear==None:
                self.front=None
            else:
                self.rear.next=None #후단삭제기 때문에 후단 노드의 next는 None이다
            return data
````
